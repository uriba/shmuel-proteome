## Load proteomic data set from csv to a DataFrame

```python
import csv
import pandas as pd
from pandas.io.parsers import read_csv

proteome_data = read_csv('Proteomics_RM_Ecoli_Calvin_Cycle_FDR_005_180814_Format.csv',encoding='iso-8859-1')
```

## Compare average CV of every protein in triplicates to total CV across samples.

```python
samples = ['WT','HF','PG-pyr','PG-pyrXyl']
triplicates = {'WT':       [u'WTa', u'WTb', u'WTc'],
               'HF':       [u'HF1a', u'HF1b', u'HF1c'],
               'PG-pyr':   [u'PG1-pyr', u'PG2-pyr', u'PG5-pyr'],
               'PG-pyrXyl':[u'PG1-pyrXyl', u'PG2-pyrXyl', u'PG5-pyrXyl']}
for samp in samples:
    proteome_data['%s-mean' % samp] = proteome_data[triplicates[samp]].mean(axis=1)
    proteome_data['%s-std' % samp] = proteome_data[triplicates[samp]].std(axis=1)
    proteome_data['%s-CV' % samp] = proteome_data['%s-std' % samp]/proteome_data['%s-mean' % samp]

all_samples = list(set().union(*triplicates.values()))

proteome_data['all-mean'] = proteome_data[all_samples].mean(axis=1)
proteome_data['all-std'] = proteome_data[all_samples].std(axis=1)
proteome_data['all-CV'] = proteome_data['all-std']/proteome_data['all-mean']
proteome_data['avg-samp-CV']=proteome_data[['%s-CV' % s for s in samples]].mean(axis=1)

plot(proteome_data['all-CV'],proteome_data['avg-samp-CV'],'.')
plot(proteome_data['all-CV'],proteome_data['all-CV'])
```

```python
proteome_data['cv-ratio'] = proteome_data['all-CV']/proteome_data['avg-samp-CV']
means = ['%s-mean' % s for s in samples]
plot(proteome_data[means].max(axis=1),proteome_data['cv-ratio'],'.')
xscale('log')
ylim(0,10)
```

```python
high_cv = proteome_data[proteome_data['cv-ratio']>3]
high_abdc_cv = high_cv[proteome_data[means].max(axis=1)>10e10]
for p in high_abdc_cv['Annotation'].values:
    print p.replace('OS=Escherichia coli ','').replace('C321.deltaA','').replace('SV=1','')
```
